# Cel projektu

Celem projektu jest opracowanie **modelu uczenia maszynowego przewidującego intensywność treningu fizycznego** (w `kcal/min`) na podstawie danych pozyskanych z zegarka Garmin. 

Projekt obejmuje pełny proces analityczny: przygotowanie danych, konstrukcję cech, budowę i ocenę modeli, interpretację ich działania oraz przygotowanie środowiska umożliwiającego wykorzystanie modelu poza notebookiem.

W ramach projektu przygotowano również lekką usługę predykcyjną (`FastAPI`) oraz interfejs użytkownika, które pozwalają wykonywać predykcje w sposób intuicyjny, bez konieczności znajomości formatu danych wejściowych.

# Rozszerzenie projektu – wykorzystanie modelu poza notebookiem: API i UI

Projekt zawiera również komponent wdrożeniowy umożliwiający wykonywanie predykcji poza środowiskiem Jupyter Notebook.

- `api.py` udostępnia endpoint `/predict`, który wczytuje zapisany pipeline i zwraca przewidywaną intensywność treningu.
- interfejs użytkownika dostępny pod `/ui` umożliwia wykonywanie predykcji bez znajomości formatu JSON.

**Instrukcja uruchomienia:**
1. Upewnij się, że wszystkie zależności zostały zainstalowane.
   
2. Przejdź do głównego katalogu projektu (tam gdzie znajduje się `api.py`)

3. Uruchom serwer:
```
uvicorn api:app --reload
```

4. Otwórz przeglądarkę i wejdź na:

   http://127.0.0.1:8000/ (automatycznie przekieruje Cię na endpoint `/ui`)

5. Wybierz aktywność oraz uzupełnij selektywne parametry, aby przeprowadzić predykcję i otrzymać przewidzianą wartość intensywności treningu fizycznego.

   
API jest również dostępne pod adresem: http://127.0.0.1:8000/docs (dokumentacja Swagger, konieczna znajomość formatu JSON)
