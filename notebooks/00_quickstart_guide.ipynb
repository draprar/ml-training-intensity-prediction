# Project Goal

The goal of this project is to develop a **machine learning model that predicts physical training intensity** (in `kcal/min`) based on data collected from a Garmin watch.

The project covers the full analytical pipeline: data preparation, feature engineering, model building and evaluation, model interpretability, and setting up an environment that enables using the model outside of the notebook.

As part of the project, a lightweight prediction service (`FastAPI`) and a user interface were also developed, allowing predictions to be made intuitively, without requiring knowledge of the input data format.

# Library Installation

1. Create a virtual environment:
   
```
python -m venv .venv
```

Activate it:

**Windows**

```
.venv\Scripts\activate
```

**Linux / macOS**

```
source .venv/bin/activate
```

2. Install dependencies:

```
pip install -r requirements.txt
```

`requirements-freeze.txt` is a snapshot of the development environment and enables full replication of the experiment:


# Project Extension – Using the Model Outside the Notebook: API and UI

The project also includes a deployment component that enables predictions to be made outside of the Jupyter Notebook environment.
- `app/main.py` defines the FastAPI application with lifespan-based model loading at startup.
- The following endpoints are available:
  - `/health` – returns the model loading status (`ok` or `model_not_loaded`),
  - `/predict` – accepts input data and returns the predicted training intensity,
  - `/ui` – a user interface that allows predictions to be made without knowledge of the JSON format.

**Startup Instructions:**
1. Make sure all dependencies have been installed.

2. Navigate to the root directory of the project.

3. Start the server:
```
uvicorn app.main:app --reload
```
4. Open your browser and go to:
   http://127.0.0.1:8000/ (you will be automatically redirected to the `/ui` endpoint)

5. Select an activity and fill in the parameters to run a prediction and receive the predicted physical training intensity value.

The API is also available at: http://127.0.0.1:8000/docs (Swagger documentation, knowledge of JSON format required)
